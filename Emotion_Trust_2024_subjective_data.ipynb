```python
# final_1 = pd.DataFrame()
# final_2 = pd.DataFrame()
final = pd.DataFrame()

item_list = ['ID','Gender','Disdainful','Scornful','Contemptuous','Hostile','Resentful','Ashamed','Humiliated','Confident','Secure','Grateful','Happy','Respectful','Nervous','Anxious','Confused','Afraid','Freaked out','Lonely','Isolated','Trust_1','Trust_2','Trust_3','Trust_4','Trust_5','Trust_6','Risk_1','Risk_2','Risk_3','Risk_4']
# Loop through the files
for filename in files:

    print(filename)
    # Construct the full file path
    filepath = os.path.join(directory, filename)

    df = pd.read_csv(filepath,header =0,low_memory=False)
    col_count = len(df.columns)
    start_row = 2

    # ---- Create a subset of dataframe for positive / negative video responses-----
    df1 = df.loc[start_row:,['ID', 'Gender', 'Emotion_1', 'Emotion_2', 'Emotion_3', 'Emotion_4', 'Emotion_5',
       'Emotion_6', 'Emotion_7', 'Emotion_8', 'Emotion_9', 'Emotion_10','Emotion_11', 'Emotion_12', 'Emotion_13', 'Emotion_14', 'Emotion_15',
       'Emotion_16', 'Emotion_17', 'Emotion_18', 'Emotion_19', 'Trust_1', 'Trust_2', 'Trust_3','Trust_4', 'Trust_5', 'Trust_6', 'Risk_1', 'Risk_2', 'Risk_3', 'Risk_4']]

    df2 = df.loc[start_row:, ['ID', 'Gender', 'Emotion_1.1', 'Emotion_2.1', 'Emotion_3.1', 'Emotion_4.1', 'Emotion_5.1',
       'Emotion_6.1', 'Emotion_7.1', 'Emotion_8.1', 'Emotion_9.1', 'Emotion_10.1','Emotion_11.1', 'Emotion_12.1', 'Emotion_13.1', 'Emotion_14.1', 'Emotion_15.1',
       'Emotion_16.1', 'Emotion_17.1', 'Emotion_18.1', 'Emotion_19.1', 'Trust_1.1', 'Trust_2.1', 'Trust_3.1', 'Trust_4.1', 'Trust_5.1', 'Trust_6.1',
       'Risk_1.1', 'Risk_2.1', 'Risk_3.1', 'Risk_4.1']]

    df1.columns.values[0:] =item_list
    df2.columns.values[0:] =item_list

    # ---- Assign group label-------
    df1["Risk Perception"] = filename.split(sep='_')[0]
    df2["Risk Perception"] = filename.split(sep='_')[0]


    #---find the order of the videos and assign the automation error label-------
    if "_2_" in filename:
     # df1 = df.iloc[start_row:, np.r_[18:20,45:74]]#18:-3]
      df1["Automation Error"] = "With Errors"

      # df2 = df.iloc[start_row:, np.r_[18:20,74:col_count-3]]#18:-3]
      df2["Automation Error"] = "Without Errors"

    else:
      # df1 = df.iloc[start_row:, np.r_[18:20,45:74]]#18:-3]
      df1["Automation Error"] = "Without Errors"

      # df2 = df.iloc[start_row:,np.r_[18:20, 74:col_count-3]]#18:-3
      df2["Automation Error"] = "With Errors"


    df1.reset_index(inplace=True,drop=True)
    df2.reset_index(inplace=True,drop=True)

    #----- Combine two sets with main dataframe ------
    #final_2 = pd.concat([final_2,df1,df2],axis=0)
    final = pd.concat([final,df1,df2],axis=0)


final.shape
```

```python
from datetime import datetime

# Get today's date
today = datetime.today()

# Format today's date as a string (optional)
formatted_today = today.strftime("%m-%d")
print(formatted_today)

```

```python
#----- Reverse 4th item in the trust scale
x = final['Trust_4'].astype(int)

final['Trust_4'] = 8-x
```

```python
#------ Calculate average trust, risk and 4 emotion categories

trust = final.iloc[:,21:27].astype(int)
risk = final.iloc[:,27:31].astype(int)
final['Trust'] = trust.mean(axis=1).round(3)
final['Risk'] = risk.mean(axis=1).round(3)

# emotions
emotion = final.iloc[:,2:21].astype(int)
final['Resentfully Aversion']=emotion.iloc[:,0:7].mean(axis=1).round(3)
final['Happily Acceptance']=emotion.iloc[:,7:12].mean(axis=1).round(3)
final['Nervously Fear']=emotion.iloc[:,12:17].mean(axis=1).round(3)
final['Lonely Isolated']=emotion.iloc[:,17:19].mean(axis=1).round(3)
final['Disdainful']=emotion.iloc[:,0].round(3) # Access column 0 (first column)
final['Scornful']=emotion.iloc[:,1].round(3)
final['Contemptuous']=emotion.iloc[:,2].round(3)
final['Hostile']=emotion.iloc[:,3].round(3)
final['Resentful']=emotion.iloc[:,4].round(3)
final['Ashamed']=emotion.iloc[:,5].round(3)
final['Humiliated']=emotion.iloc[:,6].round(3)
final['Confident']=emotion.iloc[:,7].round(3)
final['Secure']=emotion.iloc[:,8].round(3)
final['Grateful']=emotion.iloc[:,9].round(3)
final['Happy']=emotion.iloc[:,10].round(3)
final['Respectful']=emotion.iloc[:,11].round(3)
final['Afraid']=emotion.iloc[:,12].round(3)
final['Freaked out']=emotion.iloc[:,13].round(3)
final['Confused']=emotion.iloc[:,14].round(3)
final['Anxious']=emotion.iloc[:,15].round(3)
final['Nervous']=emotion.iloc[:,16].round(3)
final['Lonely']=emotion.iloc[:,17].round(3)
final['Isolated']=emotion.iloc[:,18].round(3) # Access column 18 (last valid column)

final['Trust_std'] = trust.std(axis=1).round(3)
final['Risk_std'] = risk.std(axis=1).round(3)
```

```python
aovres = pg.mixed_anova(dv='Trust', between='Risk Perception', within='Automation Error',subject='ID', data=df,effsize='n2')
print(aovres.round(3))

sns.barplot(x="Risk Perception",
		y="Trust",
    hue="Automation Error",
		errorbar='se', width=0.5,palette = 'Blues',
		data=df)

#plt.savefig('Trust.png',dpi=300)

#post-hoc comparisons
posthoc_result = pg.pairwise_tests(dv='Trust', between='Risk Perception', within='Automation Error',subject='ID', data=df,parametric=True)
print(posthoc_result.round(3))
```

```python
###Not sure If I need all the extra stuff
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pingouin as pg


final['Used Text-to-Speech'] = final['ID'].astype(int) >= 34

# Reverse the 4th item in the trust scale if not already done
if 'Trust_4' in final.columns and final['Trust_4'].dtype != int:
    final['Trust_4'] = 8 - final['Trust_4'].astype(float).fillna(0).astype(int)

# Calculate average trust, risk, and 4 emotion categories if not already done
if 'Trust' not in final.columns:
    trust = final.iloc[:, 21:27].astype(float).fillna(0).astype(int)
    risk = final.iloc[:, 27:31].astype(float).fillna(0).astype(int)
    final['Trust'] = trust.mean(axis=1).round(3)
    final['Risk'] = risk.mean(axis=1).round(3)

    emotion = final.iloc[:, 2:21].astype(float).fillna(0).astype(int)
    final['Resentfully Aversion'] = emotion.iloc[:, 0:7].mean(axis=1).round(3)
    final['Happily Acceptance'] = emotion.iloc[:, 7:12].mean(axis=1).round(3)
    final['Nervously Fear'] = emotion.iloc[:, 12:17].mean(axis=1).round(3)
    final['Lonely Isolated'] = emotion.iloc[:, 17:19].mean(axis=1).round(3)
    final['Disdainful'] = emotion.iloc[:, 0].round(3)
    final['Scornful'] = emotion.iloc[:, 1].round(3)
    final['Contemptuous'] = emotion.iloc[:, 2].round(3)
    final['Hostile'] = emotion.iloc[:, 3].round(3)
    final['Resentful'] = emotion.iloc[:, 4].round(3)
    final['Ashamed'] = emotion.iloc[:, 5].round(3)
    final['Humiliated'] = emotion.iloc[:, 6].round(3)
    final['Confident'] = emotion.iloc[:, 7].round(3)
    final['Secure'] = emotion.iloc[:, 8].round(3)
    final['Grateful'] = emotion.iloc[:, 9].round(3)
    final['Happy'] = emotion.iloc[:, 10].round(3)
    final['Respectful'] = emotion.iloc[:, 11].round(3)
    final['Afraid'] = emotion.iloc[:, 12].round(3)
    final['Freaked out'] = emotion.iloc[:, 13].round(3)
    final['Confused'] = emotion.iloc[:, 14].round(3)
    final['Anxious'] = emotion.iloc[:, 15].round(3)
    final['Nervous'] = emotion.iloc[:, 16].round(3)
    final['Lonely'] = emotion.iloc[:, 17].round(3)
    final['Isolated'] = emotion.iloc[:, 18].round(3)

    final['Trust_std'] = trust.std(axis=1).round(3)
    final['Risk_std'] = risk.std(axis=1).round(3)

# Ensure 'Risk Perception' values are consistent
final.loc[final['Risk Perception'] == 'NoInf', 'Risk Perception'] = 'NoInfo'

# Separate the data for Used Text-to-Speech and Did Not Use Text-to-Speech
df_used_tts = final[final['Used Text-to-Speech']]
df_not_used_tts = final[~final['Used Text-to-Speech']]

```

Cluster Analysis: KMeans

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Assuming 'final' dataframe already exists with necessary data

# Ensure 'Risk Perception' values are consistent
final.loc[final['Risk Perception'] == 'NoInf', 'Risk Perception'] = 'NoInfo'

# Select the emotion columns
emotion_columns = ['Disdainful', 'Scornful', 'Contemptuous', 'Hostile', 'Resentful', 'Ashamed', 'Humiliated',
                   'Confident', 'Secure', 'Grateful', 'Happy', 'Respectful', 'Nervous', 'Anxious', 'Confused',
                   'Afraid', 'Freaked out', 'Lonely', 'Isolated']

# Standardize the emotion data
scaler = StandardScaler()
scaled_emotions = scaler.fit_transform(final[emotion_columns])

```

```python
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_emotions)
    inertia.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
final['Emotion_Cluster'] = kmeans.fit_predict(scaled_emotions)

# Print the cluster centers to understand the grouping of emotions
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=emotion_columns)
print(cluster_centers_df)

```
